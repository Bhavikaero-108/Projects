```python
from ipynb.fs.full.ThermoPropertiesNASA import thermo
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
Kp_1 = []
Kp_2 = []
Kp_3 = []
for T in range (300, 3100, 100):

 h2 = thermo("H2", 1)
 o2 = thermo("O2", 1)
 h2o = thermo("H2O", 1)
 n2 = thermo("N2",1)
 no = thermo("xNO",1)
 co = thermo("CO",1)
 co2 = thermo("CO2",1)
 c = thermo("C",1)

 g_h2o = h2o.h_mole(T) - (T*h2o.s_mole(T)) - ((1*(-T*h2.s_mole(T)))+(0.5*(-T*o2.s_mole(T))))
 G_1 = g_h2o
 Kp_first = math.exp(-G_1/(8.31446*T))
 Kp_1.append(Kp_first)

 g_no = no.h_mole(T) - (T*no.s_mole(T)) - ((0.5*(-T*n2.s_mole(T)))+(0.5*(-T*o2.s_mole(T))))
 G_2 = g_no
 Kp_second = math.exp(-G_2/(8.31446*T))
 Kp_2.append(Kp_second)

 g_co = co.h_mole(T) - (T*co.s_mole(T)) - ((1*(-T*c.s_mole(T)))+(0.5*(-T*o2.s_mole(T))))
 g_co2 = co2.h_mole(T) - (T*co2.s_mole(T)) - ((1*(-T*c.s_mole(T)))+(1*(-T*o2.s_mole(T))))
 G_3 = g_co2 - g_co - g_h2o
 Kp_third = math.exp(-G_3/(8.31446*T))
 Kp_3.append(Kp_third)

print(Kp_1)
print(Kp_2)
print(Kp_3)

```

```python
#plotting Kp vs 1000/T

x = Kp_1
y = Kp_2
z = Kp_3
temperature = []
for Temp in range (300, 3100, 100):
 u = 1000/Temp
 temperature.append(u)  

plt.semilogy(temperature, x, lw=3, label ='H2 + 0.5O2 ---> H2O')
plt.semilogy(temperature, z , lw=3, label ='CO + H2O ---> CO2')
plt.semilogy(temperature, y, lw=3, label ='1/2N2 + 1/2O2 ---> NO')
plt.xlabel("1000/T")
plt.ylabel("Equilibrium Constant (Kp)")
plt.grid(True)
plt.legend()
plt.show()

Temp = np.linspace(300,3000,28)
XNO = [math.sqrt((kp)*(0.78)*(0.21)) for kp in Kp_2]
print (XNO)
plt.plot(XNO,Temp,label ='XNO vs T' )
plt.xlabel("XNo")
plt.ylabel("Temp")
plt.legend()
plt.grid(True)
```
