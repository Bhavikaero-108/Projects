```python
import cantera as ct 
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python

gas = ct.Solution('gri30.yaml')
#gas = ct.Solution(thermo="IdealGas", s
#pecies=species.values())
#gas.X = {'IXC8H18':1,'O2':25,'N2':94} T_adia

Pressures = np.linspace(40,100,10)*ct.one_atm
T_adiabatic = []
phi = 1.58
for P in Pressures :
    gas.TP= 21,P
    gas.X = {"H2":1, "O2":0.5*phi}
    gas.equilibrate("HP")
    T_adiabatic.append(gas.T)
    print(T_adiabatic,P)
plt.plot(Pressures/ct.one_atm , T_adiabatic, color="pink")
plt.grid()
plt.xlabel('Pressure')
plt.ylabel('Adiabatic Flame Temp')
```
