```python


# Create a Cantera gas object for your mechanism
gas = ct.Solution("Jerzembeck.yaml")

phi = 0.9
pressure = 30
temperature = 800
CO2_Conc = np.arange(0, 1, 0.05)
species = ['NO2', 'CO', 'CO2', 'NO', 'H2O']

for specie in species:
    CO2_Conc1 = []
    list_temperature_CO2_Conc = []
    
    for phis in CO2_Conc:
        X_O2 = 1
        X_N2 = 3.76
        X_CO2 = phis
        reactants = {'IXC8H18': 1, 'O2': X_O2, 'N2': X_N2, 'CO2': X_CO2}
        
        # Set the temperature and pressure
        gas.TP = temperature, pressure
        
        # Set the equivalence ratio with CO2 as a diluent
        gas.set_equivalence_ratio(phi, "IXC8H18", "O2:1 N2:3.76", diluent="CO2", fraction="diluent:%.6f" % phis)
        
        # Equilibrate the mixture
        gas.equilibrate('HP')
        
        CO2_Conc1.append(gas[specie].X[0])
        list_temperature_CO2_Conc.append(gas.T)
    
    plt.figure(figsize=(9, 3))
    plt.subplot(131)
    plt.plot(list_temperature_CO2_Conc, CO2_Conc1)
    plt.ylabel('Mole Fraction')
    plt.xlabel('Adiabatic Temp')
    plt.grid()

plt.show()
```
