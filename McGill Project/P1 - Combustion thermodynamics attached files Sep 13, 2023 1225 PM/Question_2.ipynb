```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
import ThermoPropertiesNASA as Ext
```

```python
const_Ru = 8.314
```

```python
def Del_Gh20(T):
    thermo_o2 = Ext.thermo('O2',32)
    thermo_h2o = Ext.thermo('H2O',18)
    thermo_h2 = Ext.thermo('H2',2)
    return (thermo_h2o.h_mole(T) - T * thermo_h2o.s_mole(T) - ( 1* ( -T * thermo_h2.s_mole(T)) + 1/2 * ( -T * thermo_o2.s_mole(T) )))
Kp_list_h20 = []
Temp_list_h20 = []
for Temp in range(300,3000,28):
    #print(Del_G0(Temp))   
    Kp_H20 = np.exp(-1 * (Del_Gh20(Temp) / (const_Ru*Temp)))
    print(f'Kp of above equation at {str(Temp)} K:{Kp_H20}')
    Kp_list_h20.append(Kp_H20)
    y= 1000/Temp
    Temp_list_h20.append(y)
plt.plot(Temp_list_h20,Kp_list_h20,color='Blue')
plt.ylabel('1000/Temp')
plt.xlabel('Equillibrium Const')
plt.semilogy()
plt.show()

def Del_GhNO(T):
    thermo_o2 = Ext.thermo('O2',32)
    thermo_xNO = Ext.thermo('xNO',30)
    thermo_n2 = Ext.thermo('N2',28)
    return ((thermo_xNO.h_mole(T)) - (T * thermo_xNO.s_mole(T)) - ( 0.5 *(-T * thermo_n2.s_mole(T)) + 0.5 *( -T * thermo_o2.s_mole(T) )))
Kp_list_xNo = []
Temp_list_xNo = []
for Temp in range(300,3000,28):
    print(Del_GhNO(Temp))   
    Kp_xNo = math.exp(-1 * (Del_GhNO(Temp) / (const_Ru*Temp)))
    print(f'Kp of above equation at {str(Temp)} K:{Kp_xNo}')
    Kp_list_xNo.append(Kp_xNo)
    y= 1000/Temp
    Temp_list_xNo.append(y)

plt.plot(Temp_list_xNo,Kp_list_xNo,color='Red')
plt.ylabel('Equillibrium Const')
plt.xlabel('1000/Temp')
plt.semilogy()
plt.show()

def Del_GhCO(T):
    thermo_o2 = Ext.thermo('O2',32)
    thermo_h2o = Ext.thermo('H2O',18)
    thermo_h2 = Ext.thermo('H2',2)
    thermo_CO = Ext.thermo('CO',28)
    thermo_CO2 = Ext.thermo('CO2',44)
    thermo_C = Ext.thermo('C',12)
    Del_Gf_H2O = (thermo_h2o.h_mole(T) - T * thermo_h2o.s_mole(T) - ( 1* ( -T * thermo_h2.s_mole(T)) + 1/2 * ( -T * thermo_o2.s_mole(T) )))
    Def_Gf_CO2 = (thermo_CO2.h_mole(T)- T* thermo_CO2.s_mole(T) - (1*(-T * thermo_C.s_mole(T))+ 1/2 * (-T * thermo_o2.s_mole(T))))
    Def_Gf_CO = (thermo_CO.h_mole(T)- T*thermo_CO.s_mole(T)-(1*(-T * thermo_C.s_mole(T))+ 1 * (-T * thermo_o2.s_mole(T))))
    return (Def_Gf_CO2 - Del_Gf_H2O - Def_Gf_CO)

Kp_list_CO2 = []
Temp_list_CO2 = []
for Temp in range(300,3000,28):
    #print(Del_G0(Temp))   
    Kp_CO2 = math.exp(-1 * (Del_GhCO(Temp) / (const_Ru*Temp)))
    print(f'Kp of above equation at {str(Temp)} K:{Kp_CO2}')
    Kp_list_CO2.append(Kp_CO2)
    y= 1000/Temp
    Temp_list_CO2.append(y)
plt.plot(Temp_list_CO2,Kp_list_CO2,color='Black')
plt.ylabel('1000/Temp')
plt.xlabel('Equillibrium Const')
plt.semilogy()
plt.show()


Temp = np.linspace(300,3000,97)
XNO = [math.sqrt((kp)*(0.78)*(0.21)) for kp in Kp_list_xNo]

plt.plot(Temp,XNO)
    
```
