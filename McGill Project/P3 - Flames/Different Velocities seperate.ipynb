```python
from pathlib import Path
import cantera as ct
import matplotlib.pyplot as plt
import pandas as pd
```

```python
p = ct.one_atm  # pressure
tin_f = 300.0  # fuel inlet temperature
tin_o = 300.0  # oxidizer inlet temperature
```

```python
gas = ct.Solution('Jerzembeck.yaml')
gas.TP = gas.T, p
density = gas.density
print(f"Density : {density}")
mdot_o1 = 0.1225  # kg/m^2/s
mdot_f1 = 0.237# kg/m^2/s
comp_o = 'O2:0.21, N2:0.78, AR:0.01'
comp_f = 'IXC8H18:1'
width = 0.02
loglevel = 1
```

```python
f1 = ct.CounterflowDiffusionFlame(gas, width=width)
f1.fuel_inlet.mdot = mdot_f1
f1.fuel_inlet.X = comp_f
f1.fuel_inlet.T = tin_f
f1.oxidizer_inlet.mdot = mdot_o1
f1.oxidizer_inlet.X = comp_o
f1.oxidizer_inlet.T = tin_o
```

```python
f1.boundary_emissivities = 0.0, 0.0
# Turn radiation off
f1.radiation_enabled = False
```

```python
f1.set_refine_criteria(ratio=4, slope=0.2, curve=0.1, prune=0.04)
f1.solve(loglevel, auto=True)
f1.show()
if "native" in ct.hdf_support():
    output = Path() / "diffusion_flame_vel1.h5"
else:
    output = Path() / "diffusion_flame_vel1.yaml"
output.unlink(missing_ok=True)

f1.save(output)
f1.save('diffusion_flame_vel1.csv', basis="mole", overwrite=True)

f1.show_stats(0)

```
